```python
from google_play_scraper import reviews
from google_play_scraper import Sort
import pandas as pd
```

```python
apps = {

    "CBE": "com.combanketh.mobilebanking",

    "CBEBirr": "prod.cbe.birr",

    "Dashen": "com.dashen.dashensuperapp",

    "BOA": "com.boa.boaMobileBanking",

    "Apollo (BOA)": "com.boa.apollo",

    "Awash Bank Mobile Banking": "pegasus.project.awash.mobile.android.bundle.mobilebank",

    "Awash Birr Pro": "com.sc.awashpay",

    "Nib Bank Mobile Banking": "com.nib.NibMobileBanking",

    "EBirr": "com.safarifone.ebirr",

    "Telebirr": "cn.tydic.ethiopay" ,
    
    "Dashen Bank Mobile Banking": "com.dashen.dashensuperapp",

    "Oromia Bank Mobile Banking": "com.oromiabank.mobilebanking",

    "Coop Bank Mobile Banking": "8821741740847225624&hl=en",


}
```

```python
all_reviews = []
```

```python
from google_play_scraper import reviews, Sort

all_reviews = []

for bank, app_id in apps.items():

    print(f"Scraping reviews for {bank}...")

    try:
        result, continuation_token = reviews(
            app_id,
            country='et',
            sort=Sort.NEWEST,
            count=20000
        )

        print(f"Collected {len(result)} reviews")

        for review in result:

            all_reviews.append({
                "review_id": review["reviewId"],
                "review": review["content"],
                "rating": review["score"],
                "date": review["at"],
                "bank": bank,
                "source": "Google Play"
            })

    except Exception as e:
        print(f"Error scraping {bank}: {e}")

print("Scraping completed.")
```

```python
print("Total reviews across all apps:", len(all_reviews))
```
